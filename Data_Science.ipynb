Data Visualization

```python
#Rename experience level
df.experience_level.replace({'EN':'Entry-Level/Junior','MI':'Mid-level/Intermediate','SE':'Senior-level/Expert','EX':'Executive-level/Director'}, inplace=True)
```

```python
#Experience Level
plt.figure(figsize=(8, 6))
ax=sns.barplot(x='experience_level', y='salary_in_usd', data=df, palette='viridis', errorbar = None)
for i in ax.containers:
    ax.bar_label(i)
plt.title('Salary vs. Experience Level', fontsize=16)
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Salary (USD)', fontsize=12)
plt.tight_layout()
plt.show()
```

1.Serior Level has the highest number of postings(280), indicating a strong demand for experienced candidates.Follwed by intermediate level with 
  213 postings.
2.The number of postings for entry-level positions is 88, indicating a lower demand compared to mid-level and senior-level roles.
3.Executive level has the lowest number of postings with only 26, suggesting a very limited demand for such roles.

```python
#Salary Level
plt.figure(figsize=(10, 6))
sns.histplot(df['salary_in_usd'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Salaries (USD)', fontsize=16)
plt.xlabel('Salary (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

 1.The distribution is skewed to the right. This means that there are more salaries concentrated in the lower range, 
  and fewer salaries in the higher range.
2. The mean or median visually appears to be somewhere around $100,000 - $200,000 based on the peak of the distribution.

```python
#Job Title
plt.figure(figsize=(15,15))
plt.title('Various Job Titles in Data Science', fontsize=20, fontweight='bold')
sns.countplot(y='job_title', data=df, order=df['job_title'].value_counts().index)
plt.xlabel('Count', fontsize=15, fontweight='bold')
plt.ylabel('Job Titles', fontsize=20,fontweight='bold')
plt.show()
```

```python
#Impact of remote work
plt.figure(figsize=(10, 6))
ax= sns.barplot(x='remote_ratio', y='salary_in_usd', data=df, errorbar=None,palette='coolwarm')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Remote Work Impact on Salary', fontsize=16)
plt.xlabel('Remote Work Ratio (%)', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.tight_layout()
plt.show()
```

1. It is noticed that as the remote work ratio increases, the average salary also tends to increase.
2.The average salary for fully remote positions (100% remote work ratio) is the highest at $122,457.
3.And the average salary for positions with no remote work option is the lowest at $80,823.
4. This trend could suggest that remote work is associated with higher-paying roles.

```python
#Replacing employment type
df.employment_type.replace({'PT':'Part-time','FT':'Full-time','CT':'Contract','FL':'Freelance'}, inplace=True)
```

```python
#Employment Type
plt.figure(figsize=(10,5))

ax = sns.countplot (x='employment_type', data=df,palette='coolwarm')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Employment Type of Data Science Jobs', fontsize=18)
plt.show()
```

1.The vast majority of data science jobs are full-time positions,
 with a count of 588. This indicates a strong preference for full-time employment in the data science field.
2.A relatively small number of jobs (5) are listed as contract positions.
3.Both part-time and freelance positions are even less common, with only 10 and 4 listings respectively.
4.The dominance of full-time employment could be attributed
 to the nature of data science work, which often involves long-term projects and collaborations.

```python
plt.figure(figsize=(10,5))

ax=sns.countplot(x='employment_type', data=df, hue = 'experience_level',palette='coolwarm')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Employment Type by Experience Level', fontsize=18)
plt.show()
```

1.The preference for full-time employment seems consistent across all experience levels in the data science field.
2.This indicates that companies in the sector generally prioritize full-time hires, regardless of the candidate's experience.

```python
#Geographical Trends
top_residences = df['employee_residence'].value_counts().head(10).index
filtered_data = df[df['employee_residence'].isin(top_residences)]

plt.figure(figsize=(14, 8))
sns.boxplot(x='employee_residence', y='salary_in_usd', data=filtered_data)
plt.title('Salary Distribution by Company Location', fontsize=16)
plt.xlabel('Employee Residence', fontsize=12)
plt.ylabel('Salary (USD)', fontsize=12)
plt.tight_layout()
plt.show()
```

 Employees in the US and FR have the highest median salaries, while employees in IN and PT have the lowest median salaries. The IQR is also wider for some countries than others, indicating that there is more variation in salaries in those countries.

```python
#Job Title Analysis
top_titles = df['job_title'].value_counts().head(10).index
filtered_titles = df[df['job_title'].isin(top_titles)]

plt.figure(figsize=(14, 8))
ax=sns.barplot(y='job_title', x='salary_in_usd', data=filtered_titles, errorbar=None, palette='coolwarm')
for i in ax.containers:
    ax.bar_label(i)
plt.title('Average Salary by Job Title (Top 10)', fontsize=16)
plt.xlabel('Average Salary (USD)', fontsize=12)
plt.ylabel('Job Title', fontsize=12)
plt.tight_layout()
plt.show()
```

1.Data Architect tops the list with an average annual salary of $177,874.
2.Big Data Engineer has the lowest average salary among the top 10 at $51,974.

```python
df.company_size.replace({'L':'Large','M':'Medium','S':'Small'}, inplace=True)
```

```python
#Company Size Influence
plt.figure(figsize=(10, 6))
sns.boxplot(x='company_size', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Company Size', fontsize=16)
plt.xlabel('Company Size', fontsize=12)
plt.ylabel('Salary (USD)', fontsize=12)
plt.tight_layout()
plt.show()
```

1.Large companies show the most significant spread in salaries, with some outliers reaching very high values.

```python
#Comapny Size by experience Level
plt.figure(figsize=(10,5))
plt.title('Company Size by Experience Level', fontsize=15)
ax=sns.countplot(x='company_size', hue='experience_level', data=df, 
              order=df['company_size'].value_counts().iloc[:3].index,palette='coolwarm')
for i in ax.containers:
    ax.bar_label(i)
plt.ylabel('Count', fontsize=10)
plt.xlabel('Company Size', fontsize=10)
plt.show()
```

The chart suggests that Medium and Large companies tend to have a more diverse workforce in terms of 
experience level, while Small companies have a higher proportion of Mid-Level and Entry-Level professionals.

```python
#Salary Over Work Years
plt.figure(figsize=(10, 6))
ax=sns.lineplot(x='work_year', y='salary_in_usd', data=df, ci=None, marker='o')
plt.title('Salary Trends Over Years', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Salary (USD)', fontsize=12)
plt.xticks([2020,2021,2022])

plt.show()
```

1.In 2020, the average salary was around $95,000.
2.In 2021, there was a noticeable jump, bringing the average salary to approximately $100,000.
3.By 2022, the average salary had increased further to around $125,000.
4.The graph indicates a positive trend in average salaries over the three years.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='experience_level', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Experience Level', fontsize=16)
plt.xlabel('Experience Level', fontsize=12)
plt.ylabel('Salary (USD)', fontsize=12)
plt.tight_layout()
plt.show()
```

1.Executive level has the highest median salary.
2.Mid-Level has the lowest median salary.
3.Senior and Entry-Level fall in between.
4.The graph suggests that, on average,
higher experience levels are associated with higher salaries, but there's also greater variability in compensation at the Executive level.
